````python
from typing import Literal
from langchain_core.language_models.chat_models import BaseChatModel
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_openai import ChatOpenAI
from langchain_deepseek import ChatDeepSeek
from langchain_xai import ChatXAI


MODEL_PROVIDER_MAP = {
    "gemini-1.5-flash": ChatGoogleGenerativeAI,
    "gemini-1.5-pro": ChatGoogleGenerativeAI,
    "gemini-2.5-flash": ChatGoogleGenerativeAI,
    "gemini-2.5-pro": ChatGoogleGenerativeAI,
    "o1": ChatOpenAI,
    "o3": ChatOpenAI,
    "gpt-4o": ChatOpenAI,
    "gpt-4o-mini": ChatOpenAI,
    "gpt-4.1": ChatOpenAI,
    "gpt-4.1-mini": ChatOpenAI,
    "o4-mini": ChatOpenAI,
    "deepseek-chat": ChatDeepSeek,
    "grok-3": ChatXAI,
    "grok-3-mini": ChatXAI,
}

# The type hint is generated from the keys of our dictionary
ModelName = Literal[
    "gemini-1.5-flash", "gemini-1.5-pro", "gemini-2.5-flash",
    "gemini-2.5-pro",
    "gpt-4o", "gpt-4o-mini", "o3", "o1", "gpt-4.1", "gpt-4.1-mini", "o4-mini",
    "deepseek-chat",
    "grok-3", "grok-3-mini"
]

def get_llm(model_name: ModelName, temperature: float = 0.0) -> BaseChatModel:
    """
    Instantiates and returns a language model based on a direct mapping.
    """
    print(f"--- ⚙️  Instantiating model: {model_name} ---")

    model_class = MODEL_PROVIDER_MAP.get(model_name)

    if model_class is None:
        raise ImportError(
            f"Model '{model_name}' is not available. "
            f"Check if its provider library (e.g., langchain_xai) is installed."
        )

    if model_name in ("o1", "o4-mini"):
        return model_class(model=model_name)

    return model_class(model=model_name, temperature=temperature)
````

Matrix tool

````python
import pandas as pd

# Load the TRIZ matrix Excel file
def load_triz_matrix(file_path):
    # Read the Excel file into a Pandas DataFrame
    triz_matrix = pd.read_excel(file_path, index_col=1)

    # Clean up column names and row index to remove leading/trailing spaces, and apply case-insensitivity
    triz_matrix.columns = triz_matrix.columns.astype(str).str.strip().str.lower()
    triz_matrix.index = triz_matrix.index.astype(str).str.strip().str.lower()  # Ensure index is string before cleaning

    return triz_matrix

triz_matrix = load_triz_matrix('tools_sources/triz_matrix.xls')

@tool
def triz_contradiction_matrix(improving_feature: str, worsening_feature: str) -> str:
    """Use TRIZ contradiction matrix to get inventive principles."""
    improving_feature = improving_feature.strip().lower()
    worsening_feature = worsening_feature.strip().lower()

    try:
        # Access the cell corresponding to the given contradiction (improving, worsening)
        principles = triz_matrix.loc[improving_feature, worsening_feature]

        # Check for NaN values and handle them
        if pd.isna(principles):
            result = "No inventive principles available for this contradiction."

        result = principles

    except KeyError:
        # Handle missing parameters
        result = f"Contradiction not found for improving '{improving_feature}' and worsening '{worsening_feature}'. Please check the parameters."

    return f"Inventive principles for improving '{improving_feature}' while worsening '{worsening_feature}': {result}"
````

````python
import json

@tool
def triz_principles(principle_numbers):
    """Retrieve multiple TRIZ principles by their numbers"""

    import json  # Ensure json module is imported

    json_file = 'tools_sources/triz_principles.json'

    # Open and load the JSON file
    with open(json_file, 'r') as file:
        triz_data = json.load(file)

    # Convert principle numbers to strings and fetch their corresponding info
    result = {}
    for number in principle_numbers:
        # --- SOLUTION ---
        # First, convert the number to an integer, then to a string.
        number_str = str(int(number))

        if number_str in triz_data:
            # Use the original number (or int(number)) as the key in the result
            result[number] = triz_data[number_str]
        else:
            result[number] = f"Principle {number} not found in the TRIZ data."

    return result

@tool
def triz_features() -> str:
    """Retrieve a 39 TRIZ possible parameters from a TXT file."""
    txt_file = 'tools_sources/triz_features.txt'

    # Open and read the TXT file
    with open(txt_file, 'r') as file:
        triz_data = file.read()

    return triz_data


````

## Agents definition

````python
def create_agent_node(prompt, name, call_model_func, tools=[], model="gpt-4o", temperature=0):
    """
    Creates a customizable agent node by binding a prompt and tools to an LLM model, returning a partial function.
    """
    llm = get_llm(model_name=model, temperature=temperature)
    if tools:
        llm = llm.bind_tools(tools)

    agent = prompt | llm
    node = functools.partial(call_model_func, agent=agent, name=name, tools=tools)
    return node


def delete_messages(state):
    steps_documentation = state["steps_documentation"]
    steps_documentation.append(HumanMessage(content=state["messages"][-3].tool_calls[0]["args"]["content"], name="DocumentationSpecialist"))

    messages = state["messages"]
    if len(messages) > 3:
        messages = [RemoveMessage(id=m.id) for m in messages[1:-1]]
    return {"messages": messages,
            "steps_documentation": steps_documentation}
````

## Agents' invoking model functions

````python
from langchain.schema import AIMessage, SystemMessage

from typing import Dict, Any
import functools


def call_project_manager_model(state, agent, name, tools) -> Dict[str, Any]:
    members = ["ProjectManager", "MechanicalEngineer", "ElectricalEngineer", "ControlSystemsEngineer", "SafetyEngineer",
            "TRIZSpecialist", "OperationsSpecialist", "DocumentationSpecialist"]
    messages = state["messages"]
    steps_documentation = state["steps_documentation"]

    members.remove(name)
    members_names = ", ".join(members)

    response = agent.invoke({"members_names": members_names,
                             "messages": messages,
                             "steps_documentation":steps_documentation})

    if "next" in state and state["next"] == "ProjectManager" and response["next"] == "ProjectManager":
        next = "FINISH"
    else:
        next = response["next"]
    message = response["message"]

    return {"next": next,
            "messages": [HumanMessage(content=message, name=name)]}


def call_agent_model(state, agent, name, tools) -> Dict[str, Any]:
    messages = state["messages"]
    members = ["ProjectManager", "MechanicalEngineer", "ElectricalEngineer", "ControlSystemsEngineer", "SafetyEngineer",
            "TRIZSpecialist", "OperationsSpecialist", "DocumentationSpecialist",
            "FinalReportMaker"]
    messages = state["messages"]
    steps_documentation = state["steps_documentation"]
    members.remove(name)
    members_names = ", ".join(members)

    tools_names = ", ".join([tool.name for tool in tools])

    response = agent.invoke({"members_names": members_names,
                             "tools": tools_names,
                             "messages": messages,
                             "steps_documentation": steps_documentation})


    if  hasattr(response, 'tool_calls') and response.tool_calls:
        if response.content:
            return {"messages": [HumanMessage(content=response.content, name=name)] + [AIMessage(content='', name=name, tool_calls=response.tool_calls)]}
        else:
            return {"messages": [AIMessage(content='', name=name, tool_calls=response.tool_calls)]}

    return {"messages": [HumanMessage(content=response.content, name=name)]}
````

## Routers

````python
def supervisor_router(state: AgentState) -> str:
    return {
        'TRIZSpecialist': 'TRIZSpecialist',
        'MechanicalEngineer': 'MechanicalEngineer',
        'ElectricalEngineer': 'ElectricalEngineer',
        'ControlSystemsEngineer': 'ControlSystemsEngineer',
        'SafetyEngineer': 'SafetyEngineer',
        'SoftwareEngineer': 'SoftwareEngineer',
        'OperationsSpecialist': 'OperationsSpecialist',
        'DocumentationSpecialist': 'DocumentationSpecialist',
        'FinalReportMaker': 'FinalReportMaker',
        'ProjectManager': 'ProjectManager',
        'FINISH': "FinalReportMaker"
    }.get(state.get('next', ''), "FinalReportMaker")


def agent_router(state: AgentState, tool_key: str, fallback_agent: str):

    messages = state["messages"]
    last_message = messages[-1]

    # Check if the last message has tool calls and route accordingly
    if hasattr(last_message, 'tool_calls') and last_message.tool_calls:
        return tool_key
    return fallback_agent

triz_specialist_router = functools.partial(agent_router, tool_key="triz_specialist_tools", fallback_agent="ProjectManager")
mechanical_engineer_router = functools.partial(agent_router, tool_key="mechanical_engineer_tools", fallback_agent="ProjectManager")
electrical_engineer_router = functools.partial(agent_router, tool_key="electrical_engineer_tools", fallback_agent="ProjectManager")
control_systems_engineer_router = functools.partial(agent_router, tool_key="control_systems_engineer_tools", fallback_agent="ProjectManager")
safety_engineer_router = functools.partial(agent_router, tool_key="safety_engineer_tools", fallback_agent="ProjectManager")
software_engineer_router = functools.partial(agent_router, tool_key="software_engineer_tools", fallback_agent="ProjectManager")
operations_specialist_router = functools.partial(agent_router, tool_key="operations_specialist_tools", fallback_agent="ProjectManager")
documentation_specialist_router = functools.partial(agent_router, tool_key="documentation_specialist_tools", fallback_agent="delete_messages")
final_report_maker_router = functools.partial(agent_router, tool_key="final_report_maker_tools", fallback_agent="ProjectManager")
````

````python
from google.colab import files
# files.download(file_name)


def run_experiment(model_name: str, thread_id: str):
    """
    Runs a single experiment with a specified model and thread ID.

    Args:
        model_name: The name of the language model to use (e.g., "gpt-4o").
        thread_id: The ID for the conversation thread.
    """
    app = create_workflow_app(llm_model=model_name, temperature=0)
    experiment_dict = {"name": f"{model_name}"}

    config = {"configurable": {"thread_id": thread_id}, "recursion_limit": 250}
    input_message = HumanMessage(content="Solve the following problem: Gantry cranes find extensive application across various industries, employed to move hefty loads and dangerous substances within shipping docks, building sites, steel plants, storage facilities, and similar industrial settings. The crane should move the load fast without causing any unnecessary excessive swing at the final position. Moreover, gantry cranes which always lift excessive load may result sudden stop of the crane. The crane operators attempt to lift heavier loads at a faster pace has led to recurrent malfunctions, including overheating, and the increased peed has caused excessive swinging or swaying of the lifted load, posing a safety hazard.")

    app.invoke({"messages": [input_message]}, config, stream_mode="values")

    final_state_snapshot = app.get_state(config)
    experiment_results = {}
    if 'steps_documentation' in final_state_snapshot.values:
        for i, message in enumerate(final_state_snapshot.values['steps_documentation']):
            experiment_results[f"step_{i+1}"] = message.content

    # Combine experiment_dict with the results
    experiment_dict.update(experiment_results)

    # Add thread_id to filename for uniqueness
    file_name = f"{experiment_dict['name']}_experiment_results_thread_{thread_id}.json"

    # Save the dictionary to a JSON file
    with open(file_name, 'w') as f:
        json.dump(experiment_dict, f, indent=4)

    print(f"Experiment results saved to {file_name}")

    files.download(file_name) # Uncomment this if you want to download the file automatically

    return experiment_dict
````

````python
# prompt: write a function to get the final value from key step_X, where x is the highest number in the keys of this dictionary

def get_final_step_value(data):
  """
  Gets the value from the highest numbered key in the format "step_X".

  Args:
    data: The dictionary containing the step keys.

  Returns:
    The value associated with the highest numbered step key, or None if no
    step keys are found.
  """
  highest_step = -1
  final_key = None
  for key in data.keys():
    if key.startswith("step_"):
      try:
        step_number = int(key.split("_")[1])
        if step_number > highest_step:
          highest_step = step_number
          final_key = key
      except ValueError:
        # Ignore keys that start with 'step_' but don't have a valid number
        pass
  if final_key:
    return data[final_key]
  else:
    return None

# Example usage with the loaded results dictionary:
# final_value = get_final_step_value(results)
# print(final_value)
````

````python

# --- 5. Updated Multi-Judge Evaluation Orchestrator ---

def run_multi_judge_evaluation(
    evaluator_models: List[ModelName],
    input_problem: str,
    prediction: str
) -> dict:
    """
    Runs a suite of evaluators on a prediction using an ensemble of judge models
    and returns an averaged final evaluation.
    """
    metric_prompts = {
        # "clarity": get_clarity_evaluator(),
        # "coherence": get_coherence_evaluator(),
        # "coverage": get_coverage_evaluator(),
        # "novelty": get_novelty_evaluator(),
        # "feasibility": get_feasibility_evaluator(),
        # "triz_adherence": get_triz_adherence_evaluator(),
        "expert_solution": get_expert_solution_evaluator()
    }

    all_judges_evaluations = defaultdict(list)

    # Loop through each judge model
    for model_name in evaluator_models:
        try:
            judge_llm = get_llm(model_name)
        except (ImportError, Exception) as e:
            print(f"Could not instantiate judge model {model_name}. Skipping. Error: {e}")
            continue

        print(f"  --- Evaluating with Judge: {model_name} ---")
        # Loop through each metric for the current judge
        for metric, prompt in metric_prompts.items():
            try:
                evaluator_chain = create_evaluator(judge_llm, prompt)

                input_data = {"prediction": prediction}
                if metric == "coverage":
                    input_data["input_problem"] = input_problem

                result = evaluator_chain.invoke(input_data)
                all_judges_evaluations[metric].append(result['text'])

            except Exception as e:
                print(f"    ! Error running '{metric}' with judge '{model_name}': {e}")

    # Average the scores from all judges for each metric
    final_averaged_evaluation = {}
    for metric, evaluations in all_judges_evaluations.items():
        if not evaluations: continue
        avg_score = np.mean([e['score'] for e in evaluations if 'score' in e and e['score'] != -1])
        final_averaged_evaluation[metric] = {
            "average_score": avg_score,
            "individual_judge_evals": evaluations # Keep details for analysis
        }

    return final_averaged_evaluation


# --- 6. Main Experiment Processing Logic (Updated) ---

def load_json_content(filepath: str) -> str:
    # (Unchanged from previous version)
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            data = json.load(f)
            return data.get('content', json.dumps(data))
    except Exception as e:
        print(f"Error loading {filepath}: {e}"); return ""

def load_md_content(filepath: str) -> str:
    """
    Loads the entire content of a markdown file as a string.
    """
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: The file at {filepath} was not found.")
        return ""
    except Exception as e:
        print(f"Error loading {filepath}: {e}")
        return ""

def process_all_experiments(
    directory: str,
    agent_model_name: str,
    input_problem: str,
    evaluator_models: List[ModelName]
) -> dict:
    """
    Processes all experiment files (.md) for a given AGENT model, evaluates each run
    using an ENSEMBLE of EVALUATOR models, and aggregates the results.
    """
    print(f"\n===== Processing experiments for AGENT: {agent_model_name} =====")
    # Search for markdown files instead of json files
    filepath_pattern = os.path.join(directory, f"{agent_model_name}_*.md")
    experiment_files = glob.glob(filepath_pattern)

    if not experiment_files:
        print(f"No experiment files found for pattern: {filepath_pattern}")
        return {}

    all_runs_evaluations = []
    for filepath in experiment_files:
        print(f"\n--- Evaluating file: {filepath} ---")
        # Load the content directly from the markdown file
        solution_output = load_md_content(filepath)

        if solution_output:
            # Get the multi-judge averaged evaluation for this single run
            multi_judge_eval = run_multi_judge_evaluation(
                evaluator_models, input_problem, solution_output
            )
            all_runs_evaluations.append(multi_judge_eval)

    if not all_runs_evaluations:
        print("No evaluations were generated for the found files.")
        return {
            "agent_model_name": agent_model_name,
            "run_count": 0,
            "evaluator_models_used": evaluator_models,
            "final_average_scores": {},
            "individual_runs_evaluations": []
        }

    # Calculate the final average scores across all runs
    final_aggregated_scores = defaultdict(list)
    for run_eval in all_runs_evaluations:
        for metric, data in run_eval.items():
            final_aggregated_scores[metric].append(data['average_score'])

    final_averages = {m: np.mean(s) for m, s in final_aggregated_scores.items()}

    return {
        "agent_model_name": agent_model_name,
        "run_count": len(all_runs_evaluations),
        "evaluator_models_used": evaluator_models,
        "final_average_scores": final_averages,
        "individual_runs_evaluations": all_runs_evaluations
    }
# --- 7. Example Usage ---
````

````python
import copy
import json


def merge_evaluation_reports(main_report: dict, expert_report: dict) -> dict:
    """
    Merges the expert solution evaluation report into the main evaluation report.

    Args:
        main_report: The dictionary with standard evaluation metrics.
        expert_report: The dictionary with the expert solution alignment metric.

    Returns:
        A new dictionary with the merged results.
    """
    # Create a deep copy to avoid modifying the original dictionary in place
    merged_report = copy.deepcopy(main_report)

    # Iterate over each model evaluated in the expert report (e.g., "gpt-4o")
    for model_name, expert_data in expert_report.items():
        # Check if the model also exists in the main report to merge into
        if model_name in merged_report:
            # 1. Merge the 'final_average_scores'
            # The .update() method adds key-value pairs from the expert scores
            # to the main scores dictionary.
            merged_report[model_name]['final_average_scores'].update(expert_data['final_average_scores'])

            # 2. Merge the 'individual_runs_evaluations'
            # We assume the lists of runs correspond one-to-one.
            for i, expert_run_eval in enumerate(expert_data['individual_runs_evaluations']):
                # Ensure we don't go out of bounds if lists are different lengths
                if i < len(merged_report[model_name]['individual_runs_evaluations']):
                    # Update the specific run's dictionary with the expert evaluation metric
                    merged_report[model_name]['individual_runs_evaluations'][i].update(expert_run_eval)
        else:
            # If the model from the expert report isn't in the main one, add it.
            merged_report[model_name] = expert_data

    return merged_report

# --- Execution ---
# Call the function to merge the two reports
merged_final_report = merge_evaluation_reports(final_report, final_report_expert_solution)

# Print the merged result in a clean, readable JSON format
print(json.dumps(merged_final_report, indent=4))
````
